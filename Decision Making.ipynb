# Decision Making In Python 

## PYTHON PROVIDES THESE FOR CHEAKING STATEMEN
### 1.IF-FOR TRUE STATEMENT ONLY
### 2.IF ELSE-FOR TRUE OR FALSE STATEMENT 
### 3.IF ELIF ELSE-FOR MULTIPLE CONDITION. CHECK ONE BY ONE GIVEN CODE 
### 4.NESTED IF-FOR CHECK MULTIPLE CONDITION AT ONE TIME

# 1.IF STATEMENT

```python
Age=19
if Age>18:
     print("you can vote")
```

# 2.IF ELSE 

```python
Age=17
if Age>18:
    print("you can vote")
else:
    print("you can't vote")
```

# 3.IF ELIF ELSE

```python
Marks = [20, 70, 90, 80, 100]

for mark in Marks:
    if mark > 90:
        print(mark, "→ A Grade")
    elif mark > 80:
        print(mark, "→ B Grade")
    elif mark > 70:
        print(mark, "→ C Grade")
    else:
        print(mark, "→ Fail")
```
