```python
def greet():
    print("welcome")
greet()    
```

```python
def greet(name):
    print("welcome",name)
print("Kashish")    
```

```python
def greet(name="Sandy"):
    print("welcome",name)
greet()  
```

```python
def greet(name,age):
    print(f"Hii {name} your age is {age}")
greet("Maddy",24)    
```

```python
def add(x,y):
  return x+y
sum=add(7,8)
print (sum)
```

```python
def add(x,y):
  return x+y,x-y
a,b= add(7,8)
print (a)
print(b)
```

```python
add=lambda x,y:x+y
sum=add(8,9)
print(sum)
```
