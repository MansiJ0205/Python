# LOOP WITH CONTROL STATEMENT

### A.THERE ARE TWO TYPES OF LOOP
### 1.FOR
### 2.WHILE
### B.THERE ARE TOW TYPE OF CONTROL STATEMENT
### 1.BREAK STATEMENT
### 2.CONTINUE STATMENT


# 1.FOR LOOP

```python
Marks=[12,18,1,17,13]
```

```python
for i in Marks:
    i=i+1
    print(i)
```

```python
for i in range(8):
    print(i)
```

```python
for i in range(1,8):
    print (i)
```

```python
for i in range (1,5):
    for j in range (1,11):
        print (i*j)
```

# FOR LOOP WITH BREAK AND CONTINUE

```python
for i in range(1,6):
    if i==3:
        break
    print(i)    
```

```python
for i in range (1,6):
    if i ==3:
        continue
    print(i)    
```

# 2. WHILE LOOP

```python
count=1
while count<=5:
    print(count)
    count+=1
```

```python
i=2
while i<=6:
    print(i)
    i=i+1
```

# WHILE LOOP WITH BREAK AND CONTINUE

```python
i =1
while i<=10:
    if i==5:
        break 
    print(i)
    i=i+1
```
